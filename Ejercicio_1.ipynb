```python
import numpy as np
import tensorflow as tf

kg = np.array([1, 5, 10, 25, 50, 100], dtype=float)
lb = np.array([2.20462, 11.0231, 22.0462, 55.1155, 110.231, 220.462], dtype=float)
```

```python
oculta1 = tf.keras.layers.Dense(units=3, input_shape=[1])
oculta2 = tf.keras.layers.Dense(units=3)
salida = tf.keras.layers.Dense(units=1)

modelo = tf.keras.Sequential([oculta1, oculta2, salida])
```

```python
modelo.compile(
    optimizer=tf.keras.optimizers.Adam(0.1),
    loss='mean_squared_error'
)
```

```python
print("Entrenando...")

historial = modelo.fit(kg, lb, epochs=1000, verbose=False)

print("Fin del entrenamiento")
```

```python
import matplotlib.pyplot as plt
plt.xlabel("# Epoca")
plt.ylabel("Magnitud de pérdida")
plt.plot(historial.history["loss"])
```

```python
resultado = modelo.predict(np.array([150]))
print(f"El peso de 150kg es aproximadamente: {resultado} libras")
```
