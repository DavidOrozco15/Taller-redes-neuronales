```python
import tensorflow as tf
import numpy as np
```

```python
inversion = np.array([1, 2, 5, 10, 20, 30, 50], dtype='float')
ventas = np.array([7, 12, 27, 52, 102, 152, 252], dtype='float')
```

```python
capa_entrada = tf.keras.layers.Dense(units=10, input_shape=[1])
capa_oculta = tf.keras.layers.Dense(units=10)
capa_salida = tf.keras.layers.Dense(units=1)

modelo = tf.keras.Sequential([capa_entrada, capa_oculta, capa_salida])
```

```python
modelo.compile(
    optimizer=tf.keras.optimizers.Adam(0.01),
    loss='mean_squared_error'
)
```

```python
print("Entrenando...")
historial = modelo.fit(inversion, ventas, epochs=1000, verbose=False)
print("Fin del entrenamiento")
```

```python
inversion_proyectada = 85
prediccion = modelo.predict(np.array([inversion_proyectada], dtype='float'))

print(f"\nPara una inversión de {inversion_proyectada} mil dólares:")
print(f"El pronóstico de ventas es de: {prediccion[0][0]:.2f} mil dólares")
```
